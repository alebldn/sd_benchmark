```python
import numpy             as np
import matplotlib.pyplot as plt
import pandas            as pd 
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
```

```python
sd_jwt_color = '#B84631'
hmac_color = '#EB9FC0'
merkle_color = '#F7B660'
monoclaim_color = '#AE9ED4'
bbs_plus_color = '#666666'
ps_color = '#7C9DC7'
cl_color = '#B2C6CD'
clpok_color = '#BE9D75'
```

```python
sd_jwt_marker = 'D'
hmac_marker = 's'
merkle_marker = 'o'
monoclaim_marker = '*'
bbs_plus_marker = 'x'
ps_marker = 'v'
cl_marker = 'p'
clpok_marker = '.'

marker_range_10 = range(0,10,1)
marker_range_100 = range(0,100,10)
```

```python
df = pd.read_csv("csv/vc_issuance_duration.csv")
print("Contents in csv file:\n", df)

fig, ax = plt.subplots()
window_size = 100
x = range(1,101,1)

ax.plot(x, df['SD-JWT'], label='SD-JWT', color=sd_jwt_color, marker=sd_jwt_marker, markevery=marker_range_100)
ax.plot(x, df['HMAC'], label='HMAC', color=hmac_color, marker=hmac_marker, markevery=marker_range_100)
ax.plot(x, df['MERKLE'], label='MERKLE', color=merkle_color, marker=merkle_marker, markevery=marker_range_100)
ax.plot(x, df['MONOCLAIM'], label='MONOCLAIM', color=monoclaim_color, marker=monoclaim_marker, markevery=marker_range_100)
ax.plot(x, df['BBS+'], label='BBS+', color=bbs_plus_color, marker=bbs_plus_marker, markevery=marker_range_100)
ax.plot(x, df['PS'], label='PS', color=ps_color, marker=ps_marker, markevery=marker_range_100)
ax.plot(x, df['CL03'], label='CL03', color=cl_color, marker=cl_marker, markevery=marker_range_100)

plt.tight_layout()

ax.set_yscale("log")
ax.xaxis.set_major_locator(MultipleLocator(10))
ax.xaxis.set_minor_locator(AutoMinorLocator(10))

ax.set_xlabel("Number of Claims")
ax.set_ylabel("Latency (us)")

ax.grid(which='major', color='#EEEEEE', linestyle='solid')
ax.grid(which='minor', color='#EEEEEE', linestyle='solid')

ax.legend(loc='upper left', fancybox=True, framealpha=0.8, prop={'size': 8})

plt.xlim(0, 100)
plt.ylim(1e1, 1e5)

plt.savefig("VC Generation Time.pdf", dpi=1000, format='pdf', bbox_inches='tight')
plt.show()
```
